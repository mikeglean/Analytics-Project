#### 4.2 Removing Unsuitable Data<a class="anchor" id="null"></a>
I discovered some zero (null) data in the TotalSteps column of the daily activity dataset during the checking and cleaning process. As a result, I decided to check and remove the zero values. I created a new table and named it daily activity new to preserve the previous dataset.

**Commands used:** SELECT, FROM, WHERE, GROUP BY, ORDER BY, CREATE TABLE, DELETE, AS

```python
#Check IF total steps = 0 IN dailyActivity TABLE
SELECT
  Id,
  COUNT(*) AS num_of_zero_steps
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_merged`
WHERE
  TotalSteps = 0
GROUP BY
  Id
ORDER BY
  num_of_zero_steps 
    
--Result: 15 ids WITH zero total steps 
    
#CREATE NEW daily activity TABLE
CREATE TABLE
  `graceful-goods-368513.Google_Capstone.dailyActivity_new` AS
SELECT
  *
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_merged`

# DELETE ALL ROWS that contain zero total steps
DELETE
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_new`
WHERE
  TotalSteps = 0 
  
--Result: 77 DATA rows deleted
```

First, I wanted to segment the users based on how frequently they wore their FitBit Fitness Tracker. I created three different user types:
* Vigorous User → wear their tracker for 21 – 31 days
* Moderate User → wear their tracker for 11 – 20 days
* Gentle User → wear their tracker for 1 -10 days

**Commands used:** SELECT, FROM, WHERE, COUNT, CASE, WHEN, GROUP BY, ORDER BY

```python
#User Type Distribution
SELECT
  Id,
  COUNT(Id) AS total_user_activity,
  CASE
    WHEN COUNT(Id) BETWEEN 21 AND 31 THEN 'Vigorous User'
    WHEN COUNT(Id) BETWEEN 11
  AND 20 THEN 'Moderate User'
    WHEN COUNT(Id) BETWEEN 1 AND 10 THEN 'Gentle User'
END
  AS user_activity_class
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_new`
GROUP BY
  Id
ORDER BY
  total_user_activity
```

**Result: More than 70% of users used Fitbit Fitness Tracker on a regular basis - between 21 and 31 days.**

<img src="https://i.imgur.com/ut6U9Ar.png" alt="user_distribution" width="600"/>

Tyron (2003) stated in his journal that taking steps every day could be an ideal parameter for gathering data about our physical activity. As a result, we will include total steps data in our analysis.

Tudor-Locke and Bassett (2004) proposed a classification based on steps per day, such as:

* Sedentary Lifestyle → total steps per day <5000
* Physically Inactive → total steps per day 5000 - 7499
* Moderately Active → total steps per day 7500 - 9999
* Physically Active → total steps per day 10000 – 12499
* Physically Active → total steps per day >= 12500

**Commands used:** SELECT,AVG, FROM, WHERE, COUNT, CASE, WHEN, GROUP BY, ORDER BY, AS

```python
SELECT
  Id,
  AVG(TotalSteps) AS avg_step,
  CASE
    WHEN AVG(TotalSteps) < 5000 THEN 'Sedentary Lifestyle'
    WHEN AVG(TotalSteps) BETWEEN 5000
  AND 7499 THEN 'Physically Inactive'
    WHEN AVG(TotalSteps) BETWEEN 7500 AND 9999 THEN 'Moderately Active'
    WHEN AVG(TotalSteps) BETWEEN 10000
  AND 12499 THEN 'Physically Active'
    WHEN AVG(TotalSteps) >= 12500 THEN 'Very Active'
END
  AS user_step_class
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_new`
GROUP BY
  Id
ORDER BY
  avg_step
```

**Result: Almost 50-50 Ratio of users who used Fitbit Fitness tracker has Active and Inactive Lifestyle. (No significant Difference)**

#### 5.2 Data Correlation <a class="anchor" id="correlation"></a> 

> **Correlation Analysis** is a statistical method used to determine whether or not there is a **relationship between two variables/datasets** and the strength of that relationship.

> Correlation analysis is used in **market research** to analyze quantitative data gathered from research methods such as surveys and polls to determine whether there are any significant connections, patterns, or trends between the two.
> 
<img src="https://i.imgur.com/SJoawdS.png" alt="avg_active_minutes" width="700"/> <img src="https://i.imgur.com/QNaFuQU.png" alt="avg_active_minutes" width="700"/>

*Checking the correlation as well with Sleep data, I created a new table by joining the activity table and the sleep table.*

**Commands used:** CREATE TABLE, SELECT,AVG, FROM, INNER JOIN, GROUP BY, AS

```python
CREATE TABLE
  `graceful-goods-368513.Google_Capstone.calories_step_sleep` AS
SELECT
  activity.Id,
  AVG(Calories) AS avg_calories,
  AVG(activity.TotalSteps) AS avg_total_step,
  AVG(sleep.TotalMinutesAsleep) AS avg_total_minutes_asleep
FROM
  `graceful-goods-368513.Google_Capstone.dailyActivity_new` AS activity
INNER JOIN
  `graceful-goods-368513.Google_Capstone.sleepDay_merged` AS sleep
ON
  activity.Id = sleep.Id
GROUP BY
  activity.Id
```
